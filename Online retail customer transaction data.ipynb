```python
df
```

```python
def categorize_time(Hour):
    if 5 <= Hour < 12:
        return 'Morning'
    elif 12 <= Hour < 17:
        return 'Afternoon'
    elif 17 <= Hour < 21:
        return 'Evening'
    else:
        return 'Night'

df["TimeOfDay"] = df["Hour"].apply(categorize_time)
```

```python
df
```

```python
df.drop("Unnamed: 0",axis = 1, inplace = True)
```

```python
df.drop("date",axis = 1, inplace = True)
```

```python
df["UnitPrice"].max()
```

```python
Q1 = df['UnitPrice'].quantile(0.25)
Q3 = df['UnitPrice'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(lower_bound,"aaaaaaaaaaa",upper_bound)
# Identify outliers
outliers = df[(df['UnitPrice'] < lower_bound) | (df['UnitPrice'] > upper_bound)]
```
